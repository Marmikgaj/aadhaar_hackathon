# Merge Aadhar Biometric CSV Files
This notebook merges all 4 CSV files from the `api_data_aadhar_biometric` directory.

```python
import pandas as pd

df1 = pd.read_csv('Data/Raw_Data/api_data_aadhar_biometric/api_data_aadhar_biometric_0_500000.csv')
df2 = pd.read_csv("Data/Raw_Data/api_data_aadhar_biometric/api_data_aadhar_biometric_500000_1000000.csv")
df3 = pd.read_csv("Data/Raw_Data/api_data_aadhar_biometric/api_data_aadhar_biometric_1000000_1500000.csv")
df4 = pd.read_csv("Data/Raw_Data/api_data_aadhar_biometric/api_data_aadhar_biometric_1500000_1861108.csv")

combined_df = pd.concat([df1, df2, df3, df4], ignore_index=True)

combined_df.info()
```

```python
combined_df.to_csv(
    "api_data_aadhar_biometric_combined.csv",
    index=False
)
```

## Demographic Data

```python
df1 = pd.read_csv('Data/Raw_Data/api_data_aadhar_demographic/api_data_aadhar_demographic_0_500000.csv')
df2 = pd.read_csv("Data/Raw_Data/api_data_aadhar_demographic/api_data_aadhar_demographic_500000_1000000.csv")
df3 = pd.read_csv("Data/Raw_Data/api_data_aadhar_demographic/api_data_aadhar_demographic_1000000_1500000.csv")
df4 = pd.read_csv("Data/Raw_Data/api_data_aadhar_demographic/api_data_aadhar_demographic_1500000_2000000.csv")
df5 = pd.read_csv("Data/Raw_Data/api_data_aadhar_demographic/api_data_aadhar_demographic_2000000_2071700.csv")
combined_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

combined_df.info()
```

```python
combined_df.to_csv(
    "api_data_aadhar_demographic_combined.csv",
    index=False
)
```
